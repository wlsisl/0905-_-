## **DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환**

사이킷런의 많은 API는 dataframe을 인자로 입력 받을 수 있지만, 기본적으로 넘파이 ndarray를 입력 인자로 사용하는 경우가 대부분이다. 

* 넘파이 ndarray,리스트, 딕셔너리를 dataframe으로 변환하기
  1. dataframe은 리스트와 넘파이 ndarray와 다르게 칼럼명을 가지고 있다. 그래서 데이터 핸들링이 편하다. 그니까 칼럼명을 지정해주자!
  2. 판다스 dataframe객체의 생성인자 data는 리스트나 딕셔너리 또는 넘파이 ndarray를 입력받는다.
  3. 생성인자 colunms는 칼럼명 리스트를 입력 받아서 생성한다.

### **1차원 형태의 데이터를 기반으로 dataframe을 생성**

1차원 형태의 데이터를 기반으로 dataframe을 생성하므로 칼럼명이 한개만 필요하다!

```python
import numpy as np

col_name1=['col1']
list1=[1,2,3]
array1=np.array(list1)
print('array1 shape:',array1.shape)

#리스트를 이용해 dataframe생성.
df_list1=pd.DataFrame(list1, columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)
#넘파이 ndarray를 이용해 DataFrame 생성.
df_array1=pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)

```

### **2차원 dataframe을 생성하자**

```python

#3개의 칼럼명이 필요함.
col_name2=['col1','col2','col3']

#2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환.
list2=[[1,2,3],[11,12,13]]
array2=np.array(list2)
print('array2 shape:',array2.shape)
df_list2=pd.DataFrame(list2, columns=col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)
df_array2=pd.DataFrame(array2, columns=col_name2)
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

### **딕셔너리를 DataFrame으로 변환하자**

딕셔너리의 키는 칼럼명으로 변환한다. 주로 문자열.

딕셔너리의 값은 키에 해당하는 칼럼 데이터로 변환한다. 주로 리스트형태.

```python
#Key는 문자열 칼럼명으로 매핑, Value는 리스트 형(또는 ndarray) 칼럼 데이터로 매핑
dict={'col1':[1,11], 'col2':[2,22], 'col3':[3,33]}
df_dict=pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
```

### **DataFrame을 넘파이 ndarray,리스트, 딕셔너리로 변환하기**

많은 머신러닝 패키지가 기본 데이터 형으로 넘파이 ndarray를 사용한다. 데이터 핸들링은 dataframe을 이용하더라도 머신러닝 패키지의 입력인자 등에 적용하기 위해 다시 넘파이 ndarray로 변환하는 경우가 빈번하게 발생한다!

```python
#DataFrame을 ndarray로 변환
array3=df_dict.values
print('df_dict.values 타입:', type(array3), 'df_dict.values shape:', array3.shape)
print(array3)
```

```python
#DataFrame을 리스트로 변환
#values로 얻은 ndarray에 tolist()를 호출한다.
list3=df_dict.values.tolist()
print('df_dict.values.tolist()의 타입:', type(list3))
print(list3)

#DataFrame을 딕셔너리로 변환
dict3=df_dict.to_dict('list')
print('\n df_dict.to_dict()의 타입:',type(dict3))
print(dict3)
```

*   ndarray와 유사하게 단일값 반환 및 슬라이싱도 가능하다.



> ### **데이터 셀렉션 및 필터링**



넘파이의 경우 - [] 연산자 내 단일 값 추출, 슬라이싱, 팬시 인덱싱, 불린 인덱싱을 통해 데이터 추출

판다스의 경우 - `ix[] , iloc[], loc[] `연산자를 통해 동일한 작업을 수행한다.

### **(1)DataFrame의 [] 연산자**

넘파이 [] 연산자 : 행의 위치, 열의 위치, 슬라이싱 범위 등을 지정해 데이터를 가져온다.

Dataframe [] 에는 **칼럼 명 문자** 또는 **인덱스로 변환 가능한 표현식**이다.
여러개 칼럼 데이터를 추출하려면 ['','']리스트를 사용하면 된다.


```python
print('단일 칼럼 데이터 추출:\n', titanic_df['Pclass'].head(3))
print('\n여러 칼럼의 데이터 추출:\n',titanic_df[['Survived','Pclass']].head(3))
print('[]안에 숫자 index는 KeyError 오류 발생:\n', titanic_df[0])

titanic_df[0:2]
#이렇게 판다스의 인덱스 형태로 변환 가능한 표현식은 입력할 수 있다!

titanic_df[titanic_df['Pclass']==3].head(3)
#불린 인덱싱 표현도 가능하다.
```

*  DataFrame 바로 뒤의 [] 연산자는 넘파이의 []나 Series의 [] 와는 다르다.
*  DataFrame 뒤의 [] 내 입력 값은 칼럼명을 지정해 칼럼 지정 연산에 사용하거나 불린 인덱스 용도로만 사용해야 한다.
*  DataFrame[0:2]와 같은 슬라이싱 연산으로 데이터를 추출하는 방법은 사용하지 않는 것이 좋다.

### **(2)DataFrame ix[] 연산자**

 - ix[0,'Pclass'] 와 같이 행 위치 지정으로 인덱스값0, 열 위치 지정으로 칼럼명 Pclass 를 입력해 데이터를 추출한다.

 - 행에 해당하는 위치 지정은 인덱스 값을 입력해야한다.
 - 열 위치 지정은 칼럼 명뿐만 아니라 칼럼의 위치 값 지정도 가능하다.
 - 현재 판다스에서 사라지고 iloc[] 와 loc[]가 대체했다.

 

#### **명칭 기반 인덱싱과 위치 기반 인덱싱의 구분**

* 명칭 기반 인덱싱 - 칼럼의 명칭을 기반으로 위치를 지정하는 방식

### ix, iloc, loc의 문제점과 주의할 점

1. 가장 중요한 것은 명칭 기반 인덱싱과 위치 기반 인덱싱의 차이를 이해하는 것.
DataFrame의 인덱스나 칼럼명으로 데이터에 접근하는 것은 명칭 기반 인덱싱이다. 
0부터 시작하는 행, 열의 위치 좌표에만 의존하는 것이 위치기반 인덱싱

2. ix[] 는 명칭기반인덱싱과 위치기반 인덱싱 모두 적용.
DataFrame의 인덱스가 숫자형일 경우 행 위치에 오는 숫자는 위치 기반 인덱싱이 아니라 명칭기반 인덱싱의 DataFrame 인덱스를 가리킴.

3. iloc[]는 위치기반인덱싱만 가능, 따라서 행과 열 위치 값으로 정수형 값을 지정해 원하는 데이터 반환함.

4. loc[]는 명칭 기반 인덱싱만 가능. 따라서 행 위치에 DataFrame인덱스가 오며, 열 위치에는 칼럼 명을 지정해 원하는 데이터를 반환함.

5. 명칭 기반 인덱싱에서 슬라이싱을 '시작점:종료점'으로 지정할 때 시작점에서 종료점을 포함한 위치에 있는 데이터 반환함.

### apply lambda 식으로 데이터 가공

* 판다스는 apply 함수에 lambda 식을 결합해 DAtaFrame이나 Series의 레코드별로 데이터를 가공하는 기능을 제공한다. 

* 칼럼에 일괄적으로 데이터 가공을 하는 것이 속도면에서 빠르나 복잡한 데이터 가공이 필요할 경우 어쩔 수 없이 apply lambda를 사용한다.

* lambda 식은 파이썬에서 함수형 프로그래밍을 지원하기 위해 만들어짐

  `lambda x(입력인자) : x**2 (반환될 입력 인자)`

* if else를 지원하는데 if절의 경우 if식보다 반환 값을 먼저 서술해야한다. 
   즉) lambda x : if x<=15 'Child' else 'Adult'가 아니라 `lambda x :'Child' if x<=15 else 'Adult' `로

* else if 는 지원하지 않는다. 이용해야한다면 else절을 ()로 내포해 ()내에서 다시 if elsefmf wjrdydgo tkdydgksek.

```python
#함수 가정
def get_square(a):
  return a**2

print('3의 제곱은:', get_square(3))

#람다는 이러한 함수내의 처리를 한 줄 식으로 쉽게 변환해줌.
```

```python
lambda_square=lambda x : x**2
print('3의 제곱은:', lambda_square(3))
```

```python
#여러개의 값을 입력인자로 사용해야 할 경우 map() 함수를 결합한다.
a=[1,2,3]
squares=map(lambda x : x**2, a)
list(squares)
```

```python
#이제 데이터프레임에서 적용해보자.
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x : len(x))

titanic_df[['Name','Name_len']].head(3)
```

```python
#Lambda식에서 if else절을 사용해보자
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x :'Child' if x<=15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
```

```python
#if else를 지원하지 않으므로 ()내부를 이용한다.
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x :'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
#나이에 따라 세분화된 분류를 수행하는 함수 생성.
def get_category(age):
  cat = ''
  if age <= 5: cat = 'Baby'
  elif age <= 12: cat = 'Child'
  elif age <= 18: cat = 'Teenager'
  elif age <= 25: cat = 'Student'
  elif age <= 35: cat = 'Young Adult'
  elif age <= 60: cat = 'Adult'
  else : cat = 'Elderly'

  return cat

#lambda 식에 위에서 생성한 get_category()함수를 반환값으로 지정.
#get_category(x)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
