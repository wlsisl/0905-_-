## **04-1 데이터 시각화가 필요한 이유**

### **(1) 앤스콤 4분할 그래프 살펴보기**
앤스콤 4분할 그래프(Anscombe's quartet) : 데이터를 시각화하지 않고 수치만 확인할 때 발생할 수 있는 함정을 보여주기 위해 만든 그래프

- 앤스콤이 지적한 함정과 데이터 시각화의 필요성

 4분할 그래프를 구성하는 데이터 집합은 4개의 그룹으로 구성되어 있는데 모든 그룹은 **평균, 분산과 같은 수칫값이나 상관관계, 회귀선**이 같다는 특징이 있다. 그래서 **'데이터 4그룹이 모두 같을 것이다'**라고 착각할 수 있다. 하지만 각 데이터 그룹을 시각화하면 데이터 그룹이 서로 다른 데이터 패턴을 가지고 있다.

### (2) 다변량 그래프 그리기
 * 3개 이상의 변수를 사용한 그래프

> 다변량 그래프가 필요한 이유 ?

앞에서는 지불금액과 팁만을 사용하여 산점도를 그렸는데 만약 여기에 성별을 추가하여 산점도를 표현하려면? **점의 색상을 다르게** 한다.
즉 점의 크기를 다르게 하던지 해서 구분 방식을 다양하게 추가한다.

 ### **`<다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프>`**

 1) 앞의 scatter_plot 산점도 그래프에 성별을 새 변수로 추가하여 색상으로 구분하기
  * 문자열은 산점도 그래프의 색상을 지정하는 값으로 사용할 수 없다. 즉 문자열을 정수로 치환하는 함수를 만들어야한다.

```python
def recode_sex(sex):
  if sex=='Female':
    return 0
  else:
    return 1
```

2) recode_sex 메서드가 반환한 값을 데이터 프레임에 추가한다.

```python
tips['sex_color']= tips['sex'].apply(recode_sex)
```

3) 테이블당 인원 수는 점의 크기로 표현하여 그래프에 추가하기

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'], y=tips['tip'],s=tips['size']*10, c=tips['sex_color'], alpha = 0.5) #alpha 인자값 : 점의 투명도 조절
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size')
```

## **04-5 seaborn 라이브러리로그래프 스타일 설정하기**
 * set_style 메서드 : 그래프의 스타일 바꾸기
 * 실행하면 계속 지정 스타일로 바뀐 채로 나옴


```python
fig, ax=plt.subplots()
ax=sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

2) whitegrid로 스타일을 설정하여 그래프를 그린다. (그래프의 배경에 가로 줄이 생김)

```python
sns.set_style('whitegrid')
fig, ax=plt.subplots()
ax=sns.violinplot(x='time', y='total_bill', hue='sex', data=tips, split=True)
```

3) for문을 이용하여 모든 스타일을 하나씩 적용한 그래프

```python
fig = plt.figure()

seaborn_styles=['darkgrid', 'whitegrid','dark','white','ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position = idx+1
  with sns.axes_style(style):
    ax=fig.add_subplot(2,3,plot_position)
    violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
    violin.set_title(style)

fig.tight_layout()
```
